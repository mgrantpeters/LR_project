```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('processed_data/03-LR_network_visualisation/03h_spatial_colocalisation/0.7_dataframe_counts_expression_network.csv')

sns.set_theme(style="whitegrid", palette="pastel")
plt.figure(figsize=(7, 3))
sns.histplot(data=df[df['sum']<=df['sum'].quantile(q=0.95)], x="sum", binwidth=.99, color = 'grey')
sns.histplot(data=df[df['sum']>df['sum'].quantile(q=0.95)], x="sum", binwidth=.99, color = 'royalblue')
plt.xlabel('Number of genes detected in spot')
plt.savefig("plots/03-LR_network_visualisation/03h_spatial_colocalisation/0.7_histogram.png", dpi = 300)
plt.show()
```
