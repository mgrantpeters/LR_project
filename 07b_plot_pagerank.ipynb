```python
import numpy as np
import pandas as pd
import mygene
import omnipath as op
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
res = ['0.4', '0.55', '0.7']
sns.set(font_scale=1.2, style = 'whitegrid')
drug = pd.read_csv('raw_data/disease_annotation/druggablegenome.csv')
for n in range (0,len(res)):
        df = pd.read_csv('processed_data/03-LR_network_visualisation/03a_genetic_association_test/pagerank_network_%s.csv' % res[n])    # Load pagerank results for LRs
        df.rename(columns = {'Unnamed: 0':'gene', '0':'pagerank_score'}, inplace = True)
        df.reset_index(inplace = True)

        df_tgts = pd.read_csv("processed_data/hot-encoded-diseases_0.100000.csv")                                                       # Load disease v gene relationship
        df_tgts['gene']=df_tgts['genes']
        df_tgts = df_tgts.drop(columns=['genes']).set_index(['gene'])

        df['source_disease'] = ' '

        for m in range(0,np.shape(df)[0]):                                                                                               # Annotate pagerank with diseases
                print(df['gene'][m])
                if df['gene'][m] in list(df_tgts.index):
                        subset = df_tgts.loc[df['gene'][m]]
                        df['source_disease'].iloc[m] = ','.join(list(subset[subset==True].index))
                else: 
                        print('Not disease linked')
        df_drug = df.merge(drug, left_on = 'gene', right_on = 'hgnc_names', how = 'left')                                               # Add druggable genome annotation
        df_drug.replace({'N': 1.0, 'Y':1.0, np.nan:0.25}, inplace = True)
        df_drug['small_mol_druggable'] = df_drug['small_mol_druggable'].astype(float)
        df_drug['source_disease'] = df_drug['source_disease'].str.split(',')
        exploded = df_drug.explode('source_disease')                                                                                    # Split rows by disease annotation to facilitate plotting
        dis_exploded = exploded[exploded['source_disease']!=' ']
        dis_exploded.to_csv('processed_data/07-pagerank_visualisation/07b_plot_pagerank/pagerank_scores_druggablediseaseannotations_%sthr.csv' % res[n])        # Save annotated dataframe


                                                                                                                                        # Make some plots
        plt.figure(figsize=(10,6))
        ax = sns.swarmplot(x = dis_exploded['source_disease'], y = dis_exploded['index'], order=["AD","ALS","EssentialTremor","FrontotemporalDementia","LBD","PD","ProgressiveSupranuclearPalsy","AnorexiaNervosa","BipolarDisorder","MajorDepressiveDisorder","NeuroticDisorder","OCD","Schizophrenia","TouretteSyndrome","UnipolarDepression","BrainAneurysm","IntracranialHemorrhage","MigraineDisorder","MigraineWithAura","MS","NarcolepsyCataplexy","Narcolepsy","PartialEpilepsy","RestlessLeg"])
        ax.invert_yaxis()
        plt.xticks(rotation = 90)
        plt.xlabel('')
        plt.ylabel('PageRank gene ranking')
        plt.title('PageRank Gene rankings')
        plt.savefig('plots/07-pagerank_visualisation/07b_plot_pagerank/pagerank_ranks_%sthr.png' % res[n], dpi = 300, bbox_inches = 'tight')
        plt.show()

        top100 = dis_exploded[dis_exploded['index']<100]
        rgb_values = list(sns.color_palette("husl", 24))
        color_labels = ["AD","ALS","EssentialTremor","FrontotemporalDementia","LBD","PD","ProgressiveSupranuclearPalsy","AnorexiaNervosa","BipolarDisorder","MajorDepressiveDisorder","NeuroticDisorder","OCD","Schizophrenia","TouretteSyndrome","UnipolarDepression","BrainAneurysm","IntracranialHemorrhage","MigraineDisorder","MigraineWithAura","MS","NarcolepsyCataplexy","Narcolepsy","PartialEpilepsy","RestlessLeg"]
        color_map = dict(zip(color_labels, rgb_values))
        top100 = dis_exploded[dis_exploded['index']<100]
        plt.figure(figsize=(11,6))
        plt.scatter(x = np.asarray(top100['source_disease']), y = np.asarray(top100['pagerank_score']), alpha = top100['small_mol_druggable'], c = top100['source_disease'].map(color_map), s = 60)
        plt.xticks(rotation = 90)
        plt.xlabel('')
        plt.ylabel('PageRank gene score')
        plt.title('Pagerank scores of top 100 highest ranking genes')
        plt.savefig('plots/07-pagerank_visualisation/07b_plot_pagerank/pagerank_scores_top100_%sthr.png' % res[n], dpi = 300, bbox_inches = 'tight')
        plt.show()
```
