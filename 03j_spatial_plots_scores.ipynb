```python
sns.set(font_scale=2)

adata = sc.read("/nemo/lab/gandhis/home/users/grantpm/LIBD_LR/grantpm/DLPFC_Visium_LIBD/processed-data/01-tangram/nonIF/layer/Br8667_mid//ad_sp_orig.h5ad", load_images = True)
df = pd.read_csv('/nemo/lab/gandhis/home/users/grantpm/LR_project/processed_data/03-LR_network_visualisation/03h_spatial_colocalisation/0.7_dataframe_counts_expression_network.csv', index_col = 0)
adata.obs['bayesSpace_harmony_9'] = adata.obs['bayesSpace_harmony_9'].cat.rename_categories({1: 'SpD1 ~ L1', 
                                                     2: 'SpD2 ~ L1', 
                                                     3 : 'SpD3 ~ L2', 
                                                     4 : 'SpD4 ~ L5', 
                                                     5 : 'SpD5 ~ L3', 
                                                     6 : 'SpD6 ~ WM', 
                                                     7 : 'SpD7 ~ L6', 
                                                     8 : 'SpD8 ~ L4', 
                                                     9 : 'SpD9 ~ WM'})

subset = df[df.index.str.contains('Br8667_mid')]
subset['matching_index'] = list(pd.DataFrame(list(subset.index.str.split('-1-')))[0])
plt.rcParams["figure.figsize"] = (8, 8)
adata.obs['score'] = adata.obs.merge(subset.set_index('matching_index')['sum'], how = 'left', right_index = True, left_on = 'key')['sum']
sc.pl.spatial(adata, color=['score'], vmax = 'p95', show = False, cmap = 'viridis')
plt.savefig('plots/03-LR_network_visualisation/03h_spatial_colocalisation/spatial_Br8667_mid_07score.png', dpi = 300, bbox_inches = 'tight')
plt.show()

sc.pl.spatial(adata, color=['bayesSpace_harmony_9'], vmax = 'p95', size = 1.2, palette = sns.color_palette("husl", 9), show = False)
plt.savefig('plots/03-LR_network_visualisation/03h_spatial_colocalisation/spatial_Br8667_mid_layers.png', dpi = 300, bbox_inches = 'tight')
plt.show()
```
